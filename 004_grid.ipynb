```python
#Author:miano
# Grid geometry node : Creates planar geometry.

primitiveType = ("polygon","mesh","nurbs","bezier","points","polygon soup")
connectivity = ("rows","columns","rows and columns","triangles","quadrilaterals",
    "alternating triangles","reverse triangles")
orientation = ("ZX PLANE","XY PLANE","YZ PLANE")
size = [2,2] #width and height of the plane
center = [0,0,0] #position of the plane
rotate = [0,0,0]
rows = 4 #number of rows
columns = 4 #number of columns

```

```python
parameters = [primitiveType,connectivity,orientation,size,center,rotate,
             rows,columns]
for parameter in parameters:
    print(parameter)
```

```python
width = parameters[3][0]
height = parameters[3][1]
print("width:",width,"height:",height)
```
